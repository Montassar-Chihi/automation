```python
# import necessary libraries
import re
import pandas as pd
import io
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from random import randint
from time import sleep
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
genes_names = []
jinet = []
drugs= []
types = []
mechanisms = []
directs = []
scores = []
kinds = []
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--start-maximized")

driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)
driver.switch_to_default_content
url= "https://www.dgidb.org/druggable_gene_categories"
driver.get(url)
# sleep(8)
links_to_visit = driver.find_elements_by_xpath('//a[@data-bind="attr: { href: link }"]')
categories_to_extract = driver.find_elements_by_xpath('//h4[@data-bind="text: text"]')
links = []
categories = []
for l in range(len(links_to_visit)):
    links.append(links_to_visit[l].get_attribute("href"))
    categories.append(categories_to_extract[l].get_attribute("textContent").split("(")[0])
for num in range(0,len(links)):
#     print('we are now in category #',num)
    driver.get(links[num])
    genes = driver.find_elements_by_tag_name("td")
    for p in range(0,len(genes),3):
#         print('we are now in gene #',p)
        try:
            genes_names.append(genes[p].get_attribute("textContent"))
        except :
            pass
genes_names_list = [item.replace("\n","") for item in genes_names]
genes_names_list = list(set(genes_names_list))
genes_names_list
```

```python
kinds = []
go = 0
def start_script():
    try:
        chrome_options = Options()
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--start-maximized")

        driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)
        driver.switch_to_default_content

        for indice in range(go,len(genes_names_list)):
            ligne = []
            liennn = "https://www.dgidb.org/categories_search_results?genes="+genes_names_list[indice]+"#_successful"
            driver.get(liennn)
            categoriyet = driver.find_elements_by_xpath('//td')
            for mawdhou3 in range(0,len(categoriyet),2):
                ligne.append(categoriyet[mawdhou3].text)
            kinds.append(ligne)
        #     break
    except:
         handle_crash()

def handle_crash():
    go = indice
    sleep(20)  # Restarts the script after 20 seconds
    start_script()

start_script()
```

```python
indice_genes_names_list = 0
indice_drugs_and_genes = 0
indice_interactions = 0
indice_score = 0
def start_script():
    try:
        chrome_options = Options()
        chrome_options.add_argument("--headless")
        chrome_options.add_argument("--start-maximized")

        driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=chrome_options)
        driver.switch_to_default_content

        for nombre in range(indice_genes_names_list,len(genes_names_list)):
        #     print('we are now in gene :',gene_name)
            try:
                driver.get("https://www.dgidb.org/genes/"+genes_names_list[nombre]+"#_interactions")
            except :
                pass
            drugs_and_genes = driver.find_elements_by_xpath('//h3[@class="item-source"]')[1:]
            interactions = driver.find_elements_by_xpath('//div[@class="cell category"]')[2:]
            score = driver.find_elements_by_xpath('//div[@class="interaction-score"]')
            for f in range(indice_drugs_and_gens,len(drugs_and_genes)):
                if re.findall(":",drugs_and_genes[f].get_attribute("textContent")) == []:
                    lista = drugs_and_genes[f].get_attribute("textContent").split("\xa0")
        #               print(lista)
                    drugs.append(lista[0])
                    jinet.append(lista[1])
            for s in range(indice_interactions,len(interactions)-1):
                if (re.findall("Interaction\sInfo:",interactions[s+1].get_attribute("textContent")) != []) and (re.findall("Interaction\sTypes\s&\sDirectionality:",interactions[s].get_attribute("textContent")) != []):
                    interaction_type = interactions[s].get_attribute("textContent").split("\n\n\n")[2]
                    types.append(interaction_type)
                    try:
                        test_1 = interactions[s+1].get_attribute("textContent").split("\n\n\n")[2]
                        test_2 = interactions[s+1].get_attribute("textContent").split("\n\n\n")[3]
                        if re.findall("Mechanism\sof\sInteraction",test_1) != [] :
                            mechanism = interactions[s+1].get_attribute("textContent").split("\n\n\n")[2].split("\n")[1]
                            mechanisms.append(mechanism)
        #                       print(mechanism)
                        else:
                            mechanisms.append("n/a")
                        if re.findall("Direct Interaction",test_2) != [] :
                            direct = interactions[s+1].get_attribute("textContent").split("\n\n\n")[3].split("\n")[1]
                            directs.append(direct)
        #                       print(direct)
                        else:
                            directs.append("n/a")
                    except:
                        mechanisms.append("n/a")
                        directs.append("n/a")
        #               print("-----------------")
            for magic in range(indice_score,len(score)):
                scores.append(score[magic].get_attribute("textContent").split(":")[1])
        #         break
        #     break
        
def handle_crash():
    indice_genes_names_list = nombre
    indice_drugs_and_genes = f
    indice_interactions = s
    indice_score = magic
    sleep(20)  # Restarts the script after 20 seconds
    start_script()

start_script()
```
